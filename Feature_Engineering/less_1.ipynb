```python
def one_hot(data,variable,top_10):
    for label in top_10:
        data[variable+'_'+label]=np.where(data[variable]==label,1,0)
data=pd.read_csv('/home/akash/Documents/100days of coding/001_My_ML/Machine-Learning-in-90-days-master/Section 1- Python Crash Course/mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
one_hot(data,'X1',top_10=top_10)
data.head()
```
